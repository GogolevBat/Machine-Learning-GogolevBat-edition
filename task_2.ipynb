# Новые колонки

Я убрал отдельные создания колонок - решил оставить готовую функцию 

~ 16 новых колонок на основе текущих

```python
def new_columns(df):
    # делаем дату типом дата
    for index_column in range(1,11):
        df['time'+str(index_column)] = pd.to_datetime(df['time'+str(index_column)])
    
    # выходные ли это
    df["weekends"] = df["time1"].apply(lambda x: (1 if datetime.date(x).weekday() in (5, 6) else 0))
    
    # дни недели
    df["weekday"] = df["time1"].apply(lambda x:  datetime.date(x).weekday())
    
    # количество посещенных сайтов
    df['value_visited_sites'] = df[["site" + str(ind) for ind in range(1, 11)]].where(df[["site" + str(ind) for ind in range(1, 11)]] != 0).count(axis=1)

    # ночь ли это
    df["night"] = df["time1"].apply(lambda x: (0 if 7 < x.hour < 22 else 1))

    # Кол-во уникальных сайтов
    df["unique_sites"] = df[["site" + str(ind) for ind in range(1, 11)]].apply(lambda x: len(x[x != 0].unique()), axis=1)

    # кол-во уникальных сайтов из кол-ва посещенных
    df["percent_unique_sites"] = df["unique_sites"] / df["value_visited_sites"]

    # находит последнюю дату визита на сайт
    df["last_visited_time"] = datetime.fromisoformat("1970-01-01 00:00:01")
    for index_site in range(10,0,-1):
        site_name = "time" + str(index_site)
        df["last_visited_time"] = df.where(df["last_visited_time"] == datetime.fromisoformat("1970-01-01 00:00:01"), other = df["last_visited_time"], axis=0)[site_name]
    # df_mini["session_difference"] = df_mini["last_visited_time"] - df_mini["time1"]
    
    # находит последнюю дату визита на сайт по минутно
    df["last_visited_time_in_minute"] = df["last_visited_time"].apply(lambda x: x.minute + x.hour * 60)
    
    # разница между концом и началом сессии
    df["session_difference"] = df["last_visited_time"] - df["time1"]
    # разность по секундно
    df["session_difference_seconds"] = df["session_difference"].apply(lambda x: x.total_seconds())
   
   
    # последний посяшенный сайт
    df["last_visited_site"] = 0

    for index_site in range(10,0,-1):
        site_name = "site" + str(index_site)
        df["last_visited_site"] = df.where(df["last_visited_site"] == 0, other = df["last_visited_site"], axis=0)[site_name]
   
    # среднее время на сайт
    df["website_browsign_speed"] = df["value_visited_sites"] / df["session_difference"].apply(lambda x: x.total_seconds())
    from math import inf
    # были бесконечности удалим
    df['website_browsign_speed'] = df['website_browsign_speed'].apply(lambda x: 0 if x == inf else x)
    
    # день и месяц
    df['day_and_mounth'] = df["time1"].apply(lambda x: x.timetuple().tm_yday)
    
    # индекс первого сайта больше 27389 (0/1) (нет/да)
    df["flag_up_then_27389"] = df["site1"].apply(lambda x: 1 if x > 27389 else 0)
    
    return df
```

Применил мешок для создания еще колонок

```python
def new_columns_with_bag_of_sites(df):
    bag_of_sites = pd.read_csv("/Users/egoglev/PycharmProjects/MachineLearning/Homeworks/homework3/session1/bag_of_sites_for_analize.csv")
    import pickle
    with open("/Users/egoglev/PycharmProjects/MachineLearning/Homeworks/homework3/session1/site.pkl", 'rb') as f: 
        data_of_site = pickle.load(f)
    
    bag_of_sites["index_site"] = bag_of_sites["Unnamed: 0"].apply(lambda x: data_of_site[x])
    
    columns = ["site" + str(ind) for ind in range(1,11)]
    
    # Создаем словарь, где ключ это индекс сайта, а значение среднее значение посящения это сайта  
    dict_sites = {int(bag_of_sites.iloc[x]["index_site"]): int(bag_of_sites.iloc[x][columns].sum() / 10) for x in
                  range(0, len(bag_of_sites))}
    # Создадим ключ 0 со значением 0 - это поля которые не заполнены 
    dict_sites[0] = 0
    
    # получаем общий рейтинг сайта
    df["website_visit_rating"] = df[columns].apply(lambda x: sum([dict_sites[ind] for ind in x]), axis = 1)
    
    # колонка показывает средний "рейтинг" среди количества посещенных сайтов
    df["rate_website_on_value_visited_sites"] = df["website_visit_rating"] / df["value_visited_sites"]

    return df
```

```python
def ordinary_graph(df, x_inp, y_inp):
    group_df = df.groupby(x_inp)[y_inp].mean().reset_index()
    plt.figure(figsize=(4, 4))
    sns.barplot(x=x_inp, y=y_inp, data=group_df,color = "purple")
    plt.xlabel(x_inp)
    plt.ylabel(y_inp)
    plt.xticks(rotation=90)
```

```python
def graph_double(x_inp,y_inp,df):    
    plt.figure(figsize=(10, 10))
    sns.countplot(x=x_inp, hue=y_inp, data=df, palette='viridis')
    plt.ylabel(y_inp)
    plt.xlabel(x_inp)
    plt.title(x_inp + " on " + y_inp)
    
    plt.show()
```
